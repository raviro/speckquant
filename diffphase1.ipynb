```python
def keyexpansion(qc):
    '''
    implements the key expansion schedule
    '''
    qc.cx(k1[2],k1[1])
    qc.cx(k1[1],k1[2])
    qc.cx(k1[2],k1[1])
    qc.cx(k1[1],k1[0])
    qc.cx(k1[0],k1[1])
    qc.cx(k1[1],k1[0])

    qc.cx(k0[1],k1[1])
    qc.cx(k0[2],k1[2])

    qc.cx(k0[2],anc[0])
    qc.cx(k0[1],k0[2])

    qc.ccx(k1[0],k0[0],k0[1])
    qc.ccx(k1[1],k0[1],k0[2])
    qc.ccx(k1[2],k0[2],anc[0])

    qc.cx(k0[2],k1[2])
    qc.ccx(k1[1],k0[1],k0[2])
    qc.cx(k0[1],k1[1])
    qc.ccx(k1[0],k0[0],k0[1])

    qc.cx(k0[1],k0[2])

    qc.cx(k0[0],k1[0])
    qc.cx(k0[1],k1[1])
    qc.cx(k0[2],k1[2])

    qc.cx(k0[1],k0[0])
    qc.cx(k0[0],k0[1])
    qc.cx(k0[1],k0[0])
    qc.cx(k0[2],k0[1])
    qc.cx(k0[1],k0[2])
    qc.cx(k0[2],k0[1])

    qc.cx(k1[0],k0[0])
    qc.cx(k1[1],k0[1])
    qc.cx(k1[2],k0[2])

        
def keyexpinverse(qc):
    '''
    implements the inverse of the key expansion schedule
    '''
    qc.cx(k1[2],k0[2])
    qc.cx(k1[1],k0[1])
    qc.cx(k1[0],k0[0])

    qc.cx(k0[2],k0[1])
    qc.cx(k0[1],k0[2])
    qc.cx(k0[2],k0[1])
    qc.cx(k0[1],k0[0])
    qc.cx(k0[0],k0[1])
    qc.cx(k0[1],k0[0])

    qc.cx(k0[2],k1[2])
    qc.cx(k0[1],k1[1])
    qc.cx(k0[0],k1[0])

    qc.cx(k0[1],k0[2])

    qc.ccx(k1[0],k0[0],k0[1])
    qc.cx(k0[1],k1[1])
    qc.ccx(k1[1],k0[1],k0[2])
    qc.cx(k0[2],k1[2])

    qc.ccx(k1[2],k0[2],anc[0])
    qc.ccx(k1[1],k0[1],k0[2])
    qc.ccx(k1[0],k0[0],k0[1])

    qc.cx(k0[1],k0[2])
    qc.cx(k0[2],anc[0])

    qc.cx(k0[2],k1[2])
    qc.cx(k0[1],k1[1])

    qc.cx(k1[1],k1[0])
    qc.cx(k1[0],k1[1])
    qc.cx(k1[1],k1[0])
    qc.cx(k1[2],k1[1])
    qc.cx(k1[1],k1[2])
    qc.cx(k1[2],k1[1])
    

def roundupdate(qc,l1):
    '''
    implements one round update schedule
    '''
    qc.cx(l1[2],l1[1])
    qc.cx(l1[1],l1[2])
    qc.cx(l1[2],l1[1])
    qc.cx(l1[1],l1[0])
    qc.cx(l1[0],l1[1])
    qc.cx(l1[1],l1[0])

    qc.cx(l1[4],l1[1])
    qc.cx(l1[5],l1[2])
    
    qc.cx(l1[5],anc[0])
    qc.cx(l1[4],l1[5])
    
    qc.ccx(l1[0],l1[3],l1[4])
    qc.ccx(l1[1],l1[4],l1[5])
    qc.ccx(l1[2],l1[5],anc[0])
    
    qc.cx(l1[5],l1[2])
    qc.ccx(l1[1],l1[4],l1[5])
    
    qc.cx(l1[4],l1[1])
    qc.ccx(l1[0],l1[3],l1[4])
    
    qc.cx(l1[4],l1[5])
    
    qc.cx(l1[3],l1[0])
    qc.cx(l1[4],l1[1])
    qc.cx(l1[5],l1[2])
    
    qc.cx(k0[0],l1[0])
    qc.cx(k0[1],l1[1])
    qc.cx(k0[2],l1[2])

    qc.cx(l1[4],l1[3])
    qc.cx(l1[3],l1[4])
    qc.cx(l1[4],l1[3])
    qc.cx(l1[5],l1[4])
    qc.cx(l1[4],l1[5])
    qc.cx(l1[5],l1[4])

    qc.cx(l1[0],l1[3])
    qc.cx(l1[1],l1[4])
    qc.cx(l1[2],l1[5])
    
def roundupinverse(qc,l1):
    '''
    implements inverse of one round update schedule
    '''
    qc.cx(l1[2],l1[5])
    qc.cx(l1[1],l1[4])
    qc.cx(l1[0],l1[3])
    
    qc.cx(l1[5],l1[4])
    qc.cx(l1[4],l1[5])
    qc.cx(l1[5],l1[4])
    qc.cx(l1[4],l1[3])
    qc.cx(l1[3],l1[4])
    qc.cx(l1[4],l1[3])
    
    qc.cx(k0[2],l1[2])
    qc.cx(k0[1],l1[1])
    qc.cx(k0[0],l1[0])
    
    qc.cx(l1[5],l1[2])
    qc.cx(l1[4],l1[1])
    qc.cx(l1[3],l1[0])
    
    qc.cx(l1[4],l1[5])
    
    qc.ccx(l1[0],l1[3],l1[4])
    qc.cx(l1[4],l1[1])
    
    qc.ccx(l1[1],l1[4],l1[5])
    qc.cx(l1[5],l1[2])
    
    qc.ccx(l1[2],l1[5],anc[0])
    qc.ccx(l1[1],l1[4],l1[5])
    qc.ccx(l1[0],l1[3],l1[4])
    
    qc.cx(l1[4],l1[5])
    qc.cx(l1[5],anc[0])
    
    qc.cx(l1[5],l1[2])
    qc.cx(l1[4],l1[1])
    
    qc.cx(l1[1],l1[0])
    qc.cx(l1[0],l1[1])
    qc.cx(l1[1],l1[0])
    qc.cx(l1[2],l1[1])
    qc.cx(l1[1],l1[2])
    qc.cx(l1[2],l1[1])
        
def stateupdate(qc,l1,r1):   
    '''
    implements the required number of rounds of (here 6) state update schedule
    '''
    qc.x(k0[0])
    qc.x(k0[1])
    qc.x(k0[2])
    qc.x(k1[0])
    qc.x(k1[2])
    
    roundupdate(qc,l1)
    roundupdate(qc,r1)
    keyexpansion(qc)
    
    roundupdate(qc,l1)
    roundupdate(qc,r1)
    keyexpansion(qc)
    
    roundupdate(qc,l1)
    roundupdate(qc,r1)
    keyexpansion(qc)
    
    roundupdate(qc,l1)
    roundupdate(qc,r1)
    keyexpansion(qc)
    
    roundupdate(qc,l1)
    roundupdate(qc,r1)
    keyexpansion(qc)
    
    roundupdate(qc,l1)
    roundupdate(qc,r1)
    
           
def stateupinverse(qc,l1,r1):
    '''
    implements the required number of rounds of (here 6) inverse state update schedule
    '''
    roundupinverse(qc,r1)
    roundupinverse(qc,l1)
    
    keyexpinverse(qc)
    roundupinverse(qc,r1)
    roundupinverse(qc,l1)
    
    keyexpinverse(qc)
    roundupinverse(qc,r1)
    roundupinverse(qc,l1)
    
    keyexpinverse(qc)
    roundupinverse(qc,r1)
    roundupinverse(qc,l1)
    
    keyexpinverse(qc)
    roundupinverse(qc,r1)
    roundupinverse(qc,l1)
    
    keyexpinverse(qc)
    roundupinverse(qc,r1)
    roundupinverse(qc,l1)
    
    qc.x(k0[0])
    qc.x(k0[1])
    qc.x(k0[2])
    qc.x(k1[0])
    qc.x(k1[2])

def roundupinversehalf(qc,l1):
    '''
    partially decrypting the right part  
    '''
    qc.cx(l1[2],l1[5])
    qc.cx(l1[1],l1[4])
    qc.cx(l1[0],l1[3])
    
    qc.cx(l1[5],l1[4])
    qc.cx(l1[4],l1[5])
    qc.cx(l1[5],l1[4])
    qc.cx(l1[4],l1[3])
    qc.cx(l1[3],l1[4])
    qc.cx(l1[4],l1[3])
    
def roundupinversehalf1(qc,l1):
    '''
    uncomputing roundupinversehalf(qc,l1)
    '''
    
    qc.cx(l1[4],l1[3])
    qc.cx(l1[3],l1[4])
    qc.cx(l1[4],l1[3])
    qc.cx(l1[5],l1[4])
    qc.cx(l1[4],l1[5])
    qc.cx(l1[5],l1[4])
    
    qc.cx(l1[0],l1[3])
    qc.cx(l1[1],l1[4])
    qc.cx(l1[2],l1[5])
    
    
    
```

```python
def grovup(qc,l1,r1):
    '''
    prepares the initial state to be input for the grover oracle
    '''
    for i in range(n):
        qc.cx(l1[i],r1[i])
        
    for i in range(n):
        if(alpha[i]==1):
            qc.x(r1[i])
    
    stateupdate(qc,l1,r1)
    roundupinversehalf(qc,l1)
    roundupinversehalf(qc,r1)
    
    for i in range(n):
        qc.cx(l1[i],r1[i])
        
def grovinv(qc,l1,r1):
    '''
    returns the modified state after applying Grovers to the initial state
    '''
    for i in range(n):
        qc.cx(l1[i],r1[i])
        
    roundupinversehalf1(qc,r1)
    roundupinversehalf1(qc,l1)
    stateupinverse(qc,l1,r1)
       
    for i in range(n):
        if(alpha[i]==1):
            qc.x(r1[i])
            
    for i in range(n):
        qc.cx(l1[i],r1[i])
    
```
