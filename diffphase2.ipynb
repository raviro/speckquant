```python
def keyexpansion(qc):
    '''
    implements the key expansion schedule
    '''
    qc.cx(k1[2],k1[1])
    qc.cx(k1[1],k1[2])
    qc.cx(k1[2],k1[1])
    qc.cx(k1[1],k1[0])
    qc.cx(k1[0],k1[1])
    qc.cx(k1[1],k1[0])

    qc.cx(k0[1],k1[1])
    qc.cx(k0[2],k1[2])

    
    qc.cx(k0[1],k0[2])

    qc.ccx(k1[0],k0[0],k0[1])
    qc.ccx(k1[1],k0[1],k0[2])

    qc.cx(k0[2],k1[2])
    qc.ccx(k1[1],k0[1],k0[2])
    qc.cx(k0[1],k1[1])
    qc.ccx(k1[0],k0[0],k0[1])

    qc.cx(k0[1],k0[2])

    qc.cx(k0[0],k1[0])
    qc.cx(k0[1],k1[1])
    qc.cx(k0[2],k1[2])

    qc.cx(k0[1],k0[0])
    qc.cx(k0[0],k0[1])
    qc.cx(k0[1],k0[0])
    qc.cx(k0[2],k0[1])
    qc.cx(k0[1],k0[2])
    qc.cx(k0[2],k0[1])

    qc.cx(k1[0],k0[0])
    qc.cx(k1[1],k0[1])
    qc.cx(k1[2],k0[2])

def roundupdate(qc,l1):
    '''
    implements one round update schedule
    '''
    qc.cx(l1[2],l1[1])
    qc.cx(l1[1],l1[2])
    qc.cx(l1[2],l1[1])
    qc.cx(l1[1],l1[0])
    qc.cx(l1[0],l1[1])
    qc.cx(l1[1],l1[0])

    qc.cx(l1[4],l1[1])
    qc.cx(l1[5],l1[2])
    
    qc.cx(l1[4],l1[5])
    
    qc.ccx(l1[0],l1[3],l1[4])
    qc.ccx(l1[1],l1[4],l1[5])
    
    qc.cx(l1[5],l1[2])
    qc.ccx(l1[1],l1[4],l1[5])
    
    qc.cx(l1[4],l1[1])
    qc.ccx(l1[0],l1[3],l1[4])
    
    qc.cx(l1[4],l1[5])
    
    qc.cx(l1[3],l1[0])
    qc.cx(l1[4],l1[1])
    qc.cx(l1[5],l1[2])
    
    qc.cx(k0[0],l1[0])
    qc.cx(k0[1],l1[1])
    qc.cx(k0[2],l1[2])

    qc.cx(l1[4],l1[3])
    qc.cx(l1[3],l1[4])
    qc.cx(l1[4],l1[3])
    qc.cx(l1[5],l1[4])
    qc.cx(l1[4],l1[5])
    qc.cx(l1[5],l1[4])

    qc.cx(l1[0],l1[3])
    qc.cx(l1[1],l1[4])
    qc.cx(l1[2],l1[5])
        
def stateupdate(qc,l1,r1):   
    '''
    implements the required number of rounds of (here 6) state update schedule
    '''
    qc.x(k0[0])
    qc.x(k0[1])
    qc.x(k0[2])
    qc.x(k1[0])
    qc.x(k1[2])
    
    roundupdate(qc,l1)
    roundupdate(qc,r1)
    keyexpansion(qc)
    
    roundupdate(qc,l1)
    roundupdate(qc,r1)
    keyexpansion(qc)
    
    roundupdate(qc,l1)
    roundupdate(qc,r1)
    keyexpansion(qc)
    
    roundupdate(qc,l1)
    roundupdate(qc,r1)
    keyexpansion(qc)
    
    roundupdate(qc,l1)
    roundupdate(qc,r1)
    keyexpansion(qc)
    
    roundupdate(qc,l1)
    roundupdate(qc,r1)
    

def roundupinverse1(qc,l1):
    '''
    implements inverse of one round update schedule
    '''
    qc.cx(l1[2],l1[5])
    qc.cx(l1[1],l1[4])
    qc.cx(l1[0],l1[3])
    
    qc.cx(l1[5],l1[4])
    qc.cx(l1[4],l1[5])
    qc.cx(l1[5],l1[4])
    qc.cx(l1[4],l1[3])
    qc.cx(l1[3],l1[4])
    qc.cx(l1[4],l1[3])
    
    qc.cx(k[2],l1[2])
    qc.cx(k[1],l1[1])
    qc.cx(k[0],l1[0])
    
    qc.cx(l1[5],l1[2])
    qc.cx(l1[4],l1[1])
    qc.cx(l1[3],l1[0])
    
    qc.cx(l1[4],l1[5])
    
    qc.ccx(l1[0],l1[3],l1[4])
    qc.cx(l1[4],l1[1])
    
    qc.ccx(l1[1],l1[4],l1[5])
    qc.cx(l1[5],l1[2])
    
    qc.ccx(l1[1],l1[4],l1[5])
    qc.ccx(l1[0],l1[3],l1[4])
    
    qc.cx(l1[4],l1[5])
    
    qc.cx(l1[5],l1[2])
    qc.cx(l1[4],l1[1])
    
    qc.cx(l1[1],l1[0])
    qc.cx(l1[0],l1[1])
    qc.cx(l1[1],l1[0])
    qc.cx(l1[2],l1[1])
    qc.cx(l1[1],l1[2])
    qc.cx(l1[2],l1[1])
    
def roundupdate1(qc,l1):
    '''
    implements one round update schedule
    '''
    qc.cx(l1[2],l1[1])
    qc.cx(l1[1],l1[2])
    qc.cx(l1[2],l1[1])
    qc.cx(l1[1],l1[0])
    qc.cx(l1[0],l1[1])
    qc.cx(l1[1],l1[0])

    qc.cx(l1[4],l1[1])
    qc.cx(l1[5],l1[2])
    
    qc.cx(l1[4],l1[5])
    
    qc.ccx(l1[0],l1[3],l1[4])
    qc.ccx(l1[1],l1[4],l1[5])
    
    qc.cx(l1[5],l1[2])
    qc.ccx(l1[1],l1[4],l1[5])
    
    qc.cx(l1[4],l1[1])
    qc.ccx(l1[0],l1[3],l1[4])
    
    qc.cx(l1[4],l1[5])
    
    qc.cx(l1[3],l1[0])
    qc.cx(l1[4],l1[1])
    qc.cx(l1[5],l1[2])
    
    qc.cx(k[0],l1[0])
    qc.cx(k[1],l1[1])
    qc.cx(k[2],l1[2])

    qc.cx(l1[4],l1[3])
    qc.cx(l1[3],l1[4])
    qc.cx(l1[4],l1[3])
    qc.cx(l1[5],l1[4])
    qc.cx(l1[4],l1[5])
    qc.cx(l1[5],l1[4])

    qc.cx(l1[0],l1[3])
    qc.cx(l1[1],l1[4])
    qc.cx(l1[2],l1[5])    
```

```python
def grovup(qc,l1,r1):
    '''
    prepares the initial state to be input for the grover oracle
    '''
    
    roundupinverse1(qc,l1)
    roundupinverse1(qc,r1)
    
    for i in range(n):
        qc.cx(l1[i],r1[i])
        
def grovinv(qc,l1,r1):
    '''
    returns the modified state after applying Grovers to the initial state
    '''
    for i in range(n):
        qc.cx(l1[i],r1[i])
        
    roundupdate1(qc,r1)
    roundupdate1(qc,l1)
    
```

```python
from qiskit import *
%matplotlib inline
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister, Aer, execute
from qiskit.tools.visualization import circuit_drawer
from qiskit.visualization import plot_histogram
import Our_Qiskit_Functions as oq
simulator = Aer.get_backend('qasm_simulator')
import numpy as np

n=6
m=3
l = QuantumRegister(n, name = 'l')
r = QuantumRegister(n, name = 'r')
k0 = QuantumRegister(m, name ='k0')
k1 = QuantumRegister(m, name ='k1')
k = QuantumRegister(m, name = 'k')
trg = QuantumRegister(1, name = 'trg')
anc1 = QuantumRegister(n-2, name = 'anc1')
cl0 = ClassicalRegister(1)
cl1 = ClassicalRegister(1)
cl2 = ClassicalRegister(1)

qc = QuantumCircuit(l,r,k0,k1,k,trg,anc1,cl0,cl1,cl2)

beta = [0,0,0,1,1,1]

pl1 = [0,1,1,0,1,0]
pl2 = [1,1,1,1,0,1]

for i in range(n):
    if(pl1[i] == 1):
        qc.x(l[i])
        
for i in range(n):
    if(pl2[i] == 1):
        qc.x(r[i])

stateupdate(qc,l,r)
        
for i in range(m):
    qc.h(k[i])    

qc.x(trg[0])

'''
finding the value of subkey which satisfies the decryption
'''

for j in range(4):
    grovup(qc,l,r)
    oq.Grover_Oracle(beta, qc, r, trg, anc1)
    grovinv(qc,l,r)
    oq.Grover_Diffusion([0,0,0], qc, k, trg, anc1)
    

qc.measure(k[0],cl0[0])
qc.measure(k[1],cl1[0])
qc.measure(k[2],cl2[0])


job = execute(qc, simulator, shots=1024)
result = job.result()
counts = result.get_counts(qc)
print('The value of subkey: ', counts)
plot_histogram(counts)
```
